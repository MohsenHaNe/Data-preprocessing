```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer 
```

```python
data = pd.read_csv('DataPreprocessing.csv')
```

```python
data.head()
```

```python
si = SimpleImputer(
    missing_values=np.NaN,
    strategy='mean',
    copy=False
)
```

```python
x = data.iloc[:,:-1]
y = data.iloc[:,-1]
```

```python
si.fit(x.iloc[:,1:3])
```

```python
x.iloc[:,1:3]=si.transform(x.iloc[:,1:3])
```

```python
r = pd.get_dummies(x.iloc[:,0])
```

```python
x = x.join(r)
```

```python
x.drop('Region',axis=1)
```

```python
x['Brazil'] = np.where(x['Brazil']==True,1,0)
x['India'] = np.where(x['India']==True,1,0)
x['USA'] = np.where(x['USA']==True,1,0)
```
